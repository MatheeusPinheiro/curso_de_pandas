# Criando Dataframe

```python
import pandas as pd
import numpy as np
```

```python
dataFrame = pd.date_range("20221201", periods=31)
print(dataFrame)
```

```python
mesDataFrame = pd.date_range("20221201", periods=12, freq='M')

print(mesDataFrame)
```

```python
#Dataframe de números aleatorios
numeros_aleatorios = pd.DataFrame(np.random.rand(5,1))
print(numeros_aleatorios)
```

```python
#Dataframe de números aleatorios
numeros_aleatorios = pd.DataFrame(np.random.rand(15,10)* 100) 
print(numeros_aleatorios)
```
