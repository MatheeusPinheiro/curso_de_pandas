```python
import pandas as pd
```

```python
# Lendo as planilhas 
vendas_loja1 = pd.read_excel(r"Arquivos\Outer_Vendas_Loja1.xlsx")
vendas_loja2 = pd.read_excel(r"Arquivos\Outer_Vendas_Loja2.xlsx")

display(vendas_loja1)
display(vendas_loja2)
```

```python
verificando_vendas_df = pd.merge(vendas_loja1, vendas_loja2, on=['Id Vendedor'], how='outer', suffixes=(' Loja 1', ' Loja 2'))

display(verificando_vendas_df)
```

```python
tratamento_dados_df = verificando_vendas_df.dropna()
del tratamento_dados_df['Vendedor Loja 2']
display(tratamento_dados_df)
```

```python
total_vendas = tratamento_dados_df['Vendas Loja 1'] + tratamento_dados_df['Vendas Loja 2']
tratamento_dados_df.loc[:,'Total Vendas'] =  total_vendas
display(tratamento_dados_df)
```
