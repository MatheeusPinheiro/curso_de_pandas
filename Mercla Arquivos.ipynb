```python
import pandas as pd
```

```python
# Lendo o arquivo do excel e transformando em Dataframe
vendas_jan_df = pd.read_excel(r'Arquivos/Vendas_Jan.xlsx')
vendas_fev_df = pd.read_excel(r'Arquivos/Vendas_Fev.xlsx')
vendas_mar_df = pd.read_excel(r'Arquivos/Vendas_Mar.xlsx')
```

```python
display(vendas_jan_df)
display(vendas_fev_df)
display(vendas_mar_df)
```

```python

vendas_jan_df = pd.concat([vendas_jan_df, vendas_fev_df], ignore_index=True)

resumindo_dataframe = vendas_jan_df[['Vendedor', 'Data Venda', 'Total Vendas']]

display(resumindo_dataframe)
```

```python
vendas_geral_df = pd.concat([vendas_jan_df, vendas_fev_df, vendas_mar_df])

resumindo_dataframe = vendas_geral_df[['Vendedor', 'Data Venda', 'Total Vendas']]

display(resumindo_dataframe)
```

```python
venda_geral_grupos = pd.concat([vendas_jan_df, vendas_fev_df, vendas_mar_df], keys=["Janeiro", "Fevereiro", "Março"])
resumindo_dataframe = venda_geral_grupos[['Vendedor', 'Data Venda', 'Total Vendas']]

display(resumindo_dataframe)
```

```python
venda_geral_grupos = pd.concat([vendas_jan_df, vendas_fev_df, vendas_mar_df], keys=["Janeiro", "Fevereiro", "Março"])
extraindo_fevereiro = venda_geral_grupos.loc['Fevereiro']
display(extraindo_fevereiro)
```
