```python
import pandas as pd
```

```python
# Lendo as planilhas 
base_lanchonete_df = pd.read_excel(r"Arquivos\Vendas_Lanchonete_Pivot.xlsx")

display(base_lanchonete_df)
```

```python
#index = coluna que quero montar o pivot
#columns = Colunas
#values = soma 
pivot_exemplo1 = base_lanchonete_df.pivot(index='Data Venda', columns='Cliente', values='Preço com Desconto')
display(pivot_exemplo1)
```

```python
#index = coluna que quero montar o pivot
#columns = Colunas
#values = soma 
pivot_exemplo2 = base_lanchonete_df.pivot(index='Cliente', columns='Data Venda', values='Preço com Desconto')
display(pivot_exemplo2)
```
