```python
notas_alunos_dataframe = pd.DataFrame({
    "Nome": ['Ana', 'Pedro', 'João'],
    "Nota 1": [9, 7, 10],
    "Nota 2": [6, 9, 8],
    "Nota 3": [7, 5, 10],
    "Nota 4": [10, 10, 6]
})


#Alterando a nota do aluno pedro
notas_alunos_dataframe.loc[1, 'Nota 2'] = 50

#Adicionando nova coluna 
notas_alunos_dataframe["Média"] = (notas_alunos_dataframe["Nota 1"] + notas_alunos_dataframe["Nota 2"] + notas_alunos_dataframe["Nota 3"] + notas_alunos_dataframe["Nota 4"]) / 4

#Criando uma coluna com o valor padrão
notas_alunos_dataframe["Faltas"] = [2, 5, 4]
display(notas_alunos_dataframe)
```
