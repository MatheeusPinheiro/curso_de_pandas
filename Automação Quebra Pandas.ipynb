```python
import pandas as pd
```

```python
base_dados_df = pd.read_excel(r'Arquivos/Vendas_Jan.xlsx')

#Imprimir os dados
display(base_dados_df)
```

```python
#subset = Coluna que queremos remover
#keep = Instrução do qual valor vai permanecer
removendo_duplicidades = base_dados_df.drop_duplicates(subset="Vendedor", keep="first")

display(removendo_duplicidades)
```

```python
# Salvando os arquivos em csv
for vendedor in removendo_duplicidades['Vendedor']:
    vendas_funcionarios = base_dados_df.loc[base_dados_df['Vendedor'] == vendedor]
    vendas_funcionarios.to_csv(f'Arquivos\Automação\Relatorio vendas {vendedor}.csv', index=False)
```

```python
# Salvando os arquivos em excel
for vendedor in removendo_duplicidades['Vendedor']:
    vendas_funcionarios = base_dados_df.loc[base_dados_df['Vendedor'] == vendedor]
    vendas_funcionarios.to_excel(f'Arquivos\Automação\Relatorio vendas {vendedor}.xlsx', index=False)
```
