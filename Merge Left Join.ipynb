```python
import pandas as pd
```

```python
# Lendo as planilhas 
vendas_df = pd.read_excel(r"Arquivos\Vendas_LEFT_JOIN.xlsx")
vendedores_df = pd.read_excel(r"Arquivos\Vendedores_LEFT_JOIN.xlsx")

display(vendas_df)
display(vendedores_df)
```

```python
#on = igual a coluna que vamos utilizar para fazer o merge
#how = como será feito 
#left = esquerda
#merge = uni os arquivos
verificando_vendas_df = pd.merge(vendas_df, vendedores_df, on=["Id Vendedor"], how="left")

display(verificando_vendas_df)
```

```python
#on = igual a coluna que vamos utilizar para fazer o merge
#how = como será feito 
#left = esquerda
#merge = uni os arquivos
verificando_vendas_df = pd.merge(vendas_df, vendedores_df, on=["Id Vendedor"], how="left", suffixes=(" Vendas", " Checagem"))

display(verificando_vendas_df)
```

```python
#dropna = Excluir linha que tenha pelo menos um valor vazio
limpando_linhas_com_nan = verificando_vendas_df.dropna()
display(limpando_linhas_com_nan)
```
