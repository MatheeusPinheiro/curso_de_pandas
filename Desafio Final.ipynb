# Desafio
 - Após carregar os dados, deixe somente as colunas de Vendedor e Total de vendas
 - Com o GroupBy, use a coluna de vendedor para criar um resumo do vendedor 
 - Salve o dataframe como um arquivo Excel.

```python
import pandas as pd
```

```python
planilha_id = '1uxYa8NKhoPQVAO_LNqNWxyn30qn5S_qD'
```

```python
dados_df = pd.read_csv(f"https://docs.google.com/spreadsheets/d/{planilha_id}/export?format=csv")
display(dados_df)
```

```python
dados_vendedor = dados_df.drop(columns=['Produto', 'Data Venda'])
display(dados_vendedor)
```

```python

dados_vendedor['Total Vendas'] = dados_vendedor['Total Vendas'].str.replace(',', '.')
dados_vendedor['Total Vendas'] = dados_vendedor['Total Vendas'].astype(float)
```

```python
resumo_vendedor = dados_vendedor.groupby('Vendedor').sum()
display(resumo_vendedor)
```

```python
resumo_vendedor.to_excel('Arquivos/Resumo de vendas.xlsx', index=False)
```
